Add a larger object as perturbation to the second image in order to create images of type II

```python
from skimage import data, measure
import skimage as sk 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skimage.util import random_noise

L = 256 
N = 2000 # number of images

I = np.zeros((L,L,N))
IG = np.zeros((L,L,N))
IL = np.zeros((L,L,N))
IP = np.zeros((L,L,N))

v = np.random.randint(2, size=N)

for k in range(N):
  if v[k] == 0:
    I_aux = data.binary_blobs(length=L, blob_size_fraction=0.1,volume_fraction=0.1) # type I blobs
    Ipert = np.zeros((L,L))
  else:
    Iperturbation = data.binary_blobs(length=L, blob_size_fraction=0.5,volume_fraction=0.1) 
    label_image_pert, nregions_pert = measure.label(Iperturbation,return_num=True)
    if nregions_pert ==1:
      Ipert = label_image_pert==1
    else:
      Ipert = label_image_pert==np.random.randint(1,nregions_pert, size=1)

    I_aux = data.binary_blobs(length=L, blob_size_fraction=0.1,volume_fraction=0.1) + Ipert# type II blobs
  
  IP[:,:,k] = Ipert
  I[:,:,k] = I_aux
  IG[:,:,k] = random_noise(I_aux,clip=True,mode='localvar')

  label_image, nregions = measure.label(I[:,:,k],return_num=True)
  IL[:,:,k] = label_image
  props_aux = measure.regionprops_table(label_image,properties=['label','area','eccentricity','extent'])

  if k == 0:
    df = pd.DataFrame(props_aux)
  else: 
    df = df.append(pd.DataFrame(props_aux), ignore_index=True)

  df_aux = pd.DataFrame(props_aux)
  a_max = max(df_aux['area'].values)

  #print('Image number {}, number of regions {}, area of largest region {}, class {}'.format(k,nregions,a_max,v[k]))
```

#2. Train a DL model

To classify images according to the class (class=0 if there are no large blobs and class = 1 if the image contains a large blob)


Shape data and split in training and test subsets:

```python
X = np.expand_dims(IG, -1)
X = np.transpose(X, (2,1,0,3))
X.shape
y=v
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

When data is uint8, scale and convert to float

```python
#X_train = X_train / 255.
#X_test = X_test / 255.
#X_train = X_train.astype('float32')
#X_test = X_test.astype('float32')
```

```python
X_train.shape
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Model / data parameters
num_classes = 2
input_shape = (L, L, 1)

# convert class vectors to binary class matrices
y_train_cat = keras.utils.to_categorical(y_train, num_classes)
y_test_cat = keras.utils.to_categorical(y_test, num_classes)

model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(name="features"),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()

# Train model:
batch_size =64
epochs = 5

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.fit(X_train, y_train_cat, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
# Validate model
from sklearn.metrics import classification_report

label_names = ['typeI', 'typeII']

score = model.evaluate(X_test, y_test_cat, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])

y_test_pred_DL_full = np.argmax(model.predict(X_test),axis=1) # convert back from one-hot categorical encoding to integers
print(classification_report(y_test, y_test_pred_DL_full, target_names=label_names))
```

```python
import tensorflow as tf

def model_modifier(current_model):
    current_model.layers[-1].activation = tf.keras.activations.linear
    return current_model
```

```python
from tensorflow.keras.preprocessing.image import load_img

# Image titles
image_titles = ['typeI', 'typeII']

ind0 = np.where(y_test==0) # find index for the first type I blob in test (class = 0)
ind1 = np.where(y_test==1) # find index for the first type II blob in test (class = 1)


# Load images and Convert them to a Numpy array
img1 = X_test[ind0[0][0]].squeeze()
img2 = X_test[ind1[0][0]].squeeze()


fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(1, 2, 1)
ax.imshow(img1, cmap='gray') 
ax.set_title('img1: ' + image_titles[0])
ax = fig.add_subplot(1, 2, 2)
ax.imshow(img2, cmap='gray') 
ax.set_title('img2: '+ image_titles[1])
plt.show()
```

```python
from tf_keras_vis.utils.scores import CategoricalScore

images = np.asarray([np.array(img1), np.array(img2)])
X = images.reshape(2,L,L,1)

score = CategoricalScore([0,1])
```

```python
from matplotlib import cm
from tf_keras_vis.gradcam import Gradcam

# Create Gradcam object
gradcam = Gradcam(model,model_modifier=model_modifier,clone=False)

# Generate heatmap with GradCAM
cam = gradcam(score, X, penultimate_layer=-1)

fig = plt.figure(figsize=(18, 8))
ax = fig.add_subplot(2, 2, 1)
plt.imshow(cam[0,:,:])
plt.colorbar()
ax = fig.add_subplot(2, 2, 2)
plt.imshow(img1)
plt.colorbar()
ax = fig.add_subplot(2, 2, 3)
plt.imshow(img2)
plt.colorbar()
ax = fig.add_subplot(2, 2, 4)
plt.imshow(cam[1,:,:])
plt.colorbar()
plt.show()
```
