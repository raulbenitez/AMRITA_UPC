#Blob Parameters:

**blob_size_fraction**

Typical linear size of blob, as a fraction of length, should be smaller than 1.


**volume_fraction float, default 0.5**

Fraction of image pixels covered by the blobs (where the output is 1). Should be in [0, 1].

```python
from skimage import data, measure
import skimage as sk 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skimage.util import random_noise

L = 256

I1= data.binary_blobs(length=L, blob_size_fraction=0.3,volume_fraction=0.1) 
I2= data.binary_blobs(length=L, blob_size_fraction=0.3,volume_fraction=0.5) 
I3= data.binary_blobs(length=L, blob_size_fraction=0.3,volume_fraction=0.9) 

I4= data.binary_blobs(length=L, blob_size_fraction=0.1,volume_fraction=0.1) 
I5= data.binary_blobs(length=L, blob_size_fraction=0.1,volume_fraction=0.5) 
I6= data.binary_blobs(length=L, blob_size_fraction=0.1,volume_fraction=0.9) 
  
fig, axes = plt.subplots(3,2,figsize=(14, 9))
ax = axes.ravel()
ax[0].imshow(I1,cmap=plt.cm.gray)
ax[0].set_title('size_fraction = 0.3 vol_fraction = 0.1') 
ax[1].imshow(I2,cmap=plt.cm.gray)
ax[1].set_title('size_fraction = 0.3 vol_fraction = 0.5') 
ax[2].imshow(I3,cmap=plt.cm.gray)
ax[2].set_title('size_fraction = 0.3 vol_fraction = 0.9') 
ax[3].imshow(I4,cmap=plt.cm.gray)
ax[3].set_title('size_fraction = 0.1 vol_fraction = 0.1') 
ax[4].imshow(I5,cmap=plt.cm.gray)
ax[4].set_title('size_fraction = 0.1 vol_fraction = 0.5') 
ax[5].imshow(I6,cmap=plt.cm.gray)
ax[5].set_title('size_fraction = 0.1 vol_fraction = 0.9') 

ax[0].set_axis_off()
ax[1].set_axis_off()
ax[2].set_axis_off()
ax[3].set_axis_off()
ax[4].set_axis_off()
ax[5].set_axis_off()
plt.tight_layout()
plt.show()
```
